### Data Loading and Preprocessing 

```python
def get_distance(lat1, long1, lat2, long2):
    return geopy.distance.geodesic((lat1, long1), (lat2, long2)).km
```

```python
vents = pd.read_csv('HydrothermalVents.csv',low_memory=False)
vents = vents[vents['Activity'] != 'inactive']
#vents = vents[vents['Activity'] == 'active']

vents = vents[vents['Depth'].notnull()]
vents = vents[vents['Depth'] != '0']
vents = vents[vents['Depth'] != 'NotProvided']

vents
```

```python
#If a depth contains a range of values, replace it with the average of the range

cur_index = 0
indices = vents.index
for d in vents['Depth'].values:
    d = d.replace(' ', '')
    if('-' in d):
        num1 = d[:d.index('-')]
        print(num1)
        num2 = d[d.index('-')+1:]
        print(num2)

        average_d = (int(num1)+int(num2)) / 2

        vents.loc[indices[cur_index],'DepthAverage'] = int(average_d)
    else:
        print(int(d))
        vents.loc[indices[cur_index],'DepthAverage'] = int(d)
    
    cur_index = cur_index + 1
```

```python
geotraces = pd.read_csv('GEOTRACES_IDP2021_Seawater_Discrete_Sample_Data_v1.csv',low_memory=False)
geotraces.columns = [c.replace(' ', '_') for c in geotraces.columns]
geotraces.columns = [c.replace(':', '_') for c in geotraces.columns]
geotraces_iron = geotraces[geotraces['Fe_D_CONC_BOTTLE_[nmol/kg]'].notnull()]

geotraces_iron = geotraces_iron[['Station_METAVAR_INDEXED_TEXT', 'yyyy-mm-ddThh_mm_ss.sss', 'Longitude_[degrees_east]', 'Latitude_[degrees_north]', 'Fe_D_CONC_BOTTLE_[nmol/kg]']]

geotraces_iron.rename(columns={'Station_METAVAR_INDEXED_TEXT': 'StationID', 'yyyy-mm-ddThh_mm_ss.sss': 'dateTime', 'Longitude_[degrees_east]': 'lon', 'Latitude_[degrees_north]':'lat', 'Fe_D_CONC_BOTTLE_[nmol/kg]':'dFe'}, inplace=True)

geotraces_iron = geotraces_iron.drop_duplicates()
```

### Exploring Correlations

```python
def shortest_distance_to_vent():
    
    vent_lats = list(vents.latitude)
    vent_lons = list(vents.longitude)

    
    for index, row in geotraces_iron.iterrows():
        print(index)
        station_lon = row['lon']
        station_lat = row['lat']
        
        #print('Station Lat: ', station_lat)
        #print('Station Long: ', station_lon)

        smallest_distance_to_vent = 9999
        distance = 9999
        for i in range(len(vent_lats)):
           # print(vent_lats[i], vent_lons[i], station_lat, station_lon)
            distance = get_distance(vent_lats[i], vent_lons[i], station_lat, station_lon)
            if(distance < smallest_distance_to_vent):
                smallest_distance_to_vent = distance
        
        #print('Smallest Distance to Vent: ', smallest_distance_to_vent)
        geotraces_iron.loc[index, 'DistanceToNearestVent'] = smallest_distance_to_vent
    
shortest_distance_to_vent()
```

```python
geotraces_iron['dFe'].corr(geotraces_iron['DistanceToNearestVent'])
```
